```python
import pandas as pd
import json
```

```python
df = pd.read_json("epl_2022_2023_07_02_2023.json", encoding="utf-8")
df.head(10)
```

```python
df = df.T
df.head()
```

```python
team1_stats = df['team1_stat'].apply(pd.Series)
team1_stats = team1_stats.add_prefix('team1_')
team1_stats.head()
```

```python
team2_stats = df['team2_stat'].apply(pd.Series)
team2_stats = team2_stats.add_prefix('team2_')
team2_stats.head()
```

```python
df = pd.concat([df, team1_stats, team2_stats], axis=1)
df.drop(['team1_stat', 'team2_stat'], axis=1, inplace=True)

df.head()
```

```python
df = df.reset_index()
df.rename(columns={'index': 'match_id'}, inplace=True)

df.head()
```

```python
df['event_clean'] = df['event'].astype(str).str.replace('[', '', regex=False).str.replace(']', '', regex=False)
```

```python
parts = df['event_clean'].str.split(',')

df['team1_goals'] = parts.str[1].str.extract('(\d+)')
df['team2_goals'] = parts.str[2].str.extract('(\d+)')

df['team1_goals'] = pd.to_numeric(df['team1_goals'], errors='coerce')
df['team2_goals'] = pd.to_numeric(df['team2_goals'], errors='coerce')
```

```python
df[['event_clean', 'team1_goals', 'team2_goals']].head(10)
```

```python
def get_result(row):
    if row['team1_goals'] > row['team2_goals']:
        return 'Home Win'
    elif row['team1_goals'] < row['team2_goals']:
        return 'Away Win'
    else:
        return 'Draw'

df['result'] = df.apply(get_result, axis=1)
```

```python
def team1_points(row):
    if row['result'] == 'Home Win':
        return 3
    elif row['result'] == 'Draw':
        return 1
    else:
        return 0

def team2_points(row):
    if row['result'] == 'Away Win':
        return 3
    elif row['result'] == 'Draw':
        return 1
    else:
        return 0

df['team1_points'] = df.apply(team1_points, axis=1)
df['team2_points'] = df.apply(team2_points, axis=1)
```

```python
df['goal_diff'] = df['team1_goals'] - df['team2_goals']
```

```python
df[['team1_name', 'team2_name', 'team1_goals', 'team2_goals', 'result', 'team1_points', 'team2_points']].head(10)
```

```python
team1_df = df[['team1_name', 'team1_goals', 'team2_goals', 'team1_points']].copy()

team1_df.columns = ['team', 'goals_scored', 'goals_conceded', 'points']
```

```python
team2_df = df[['team2_name', 'team2_goals', 'team1_goals', 'team2_points']].copy()

team2_df.columns = ['team', 'goals_scored', 'goals_conceded', 'points']
```

```python
league_df = pd.concat([team1_df, team2_df])
```

```python
league_table = league_df.groupby('team').sum().reset_index()
```

```python
league_table = league_table.sort_values(by='points', ascending=False)
```

```python
league_table.head(10)
```

```python
league_table['goal_diff'] = league_table['goals_scored'] - league_table['goals_conceded']
```

```python
league_table.head(10)
```

```python
league_table.head(20)
```

```python
league_table = league_table.sort_values(by=['points', 'goal_diff'], ascending=False)
```

```python
league_table = league_table.reset_index(drop=True)
```

```python
league_table['rank'] = league_table.index + 1
```

```python
league_table.head(20)
```

```python
league_table.to_csv('league_table.csv', index=False)
```
